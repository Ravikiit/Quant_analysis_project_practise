```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
ind_data = pd.read_csv('Indices_Data_2.csv', index_col='Date')
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1

ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```
